```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from jupyterthemes import jtplot as jtp
```

```python
hrs_df = pd.read_csv('HomeRunData.csv')
#hr_df = hr_df.drop(columns = ['pitch_type', 'release_speed','release_pos_x',])
hits_df = pd.read_csv('HitsData.csv', low_memory=False)
hit_df = hits_df[['player_name','batter', 'events', 'launch_speed', 'launch_angle', 'game_year']]
pd.set_option('display.max_rows', 10, 'display.max_columns', 10)
hit_df
```

```python
#hr_df = hr_df.drop(columns=['pitch_type','game_date','release_speed','release_pos_x','release_pos_z','pitcher','description','home_score','away_score','bat_score','fld_score','post_away_score','post_home_score','post_bat_score','post_fld_score','if_fielding_alignment', 'of_fielding_alignment','spin_dir', 'spin_rate_deprecated','break_angle_deprecated','break_length_deprecated','zone','game_type','estimated_ba_using_speedangle','estimated_woba_using_speedangle','woba_value','woba_denom','babip_value','iso_value','launch_speed_angle','at_bat_number','pitch_number','pitch_name'])
hr_df = hrs_df[['player_name','batter', 'events', 'launch_speed', 'launch_angle', 'game_year']]
hr_df
```

```python
ab_df = pd.concat([hr_df,hit_df])
```

```python
ab_df = ab_df.loc[ab_df['game_year'] < 2020]
ab_df
```

```python
events = []

for event in ab_df['events']:
    if event == 'home_run':
        events.append('home_run')
    else:
        events.append('base_hit')
        
events
```

```python
ab_df['event_binary'] = events
ab_df = ab_df.drop(columns=['events', 'player_name','batter'])
ab_df
```

```python
ab_df = ab_df.dropna()
ab_df
```

```python
hr= ab_df.loc[ab_df['event_binary']=='home_run']
xhr=hr['launch_speed']
yhr=hr['launch_angle']
hit=ab_df.loc[ab_df['event_binary']=='base_hit']
xhit = hit['launch_speed']
yhit = hit['launch_angle']

# for graders comment out line 9 "jtp.style(theme='onedork')"
jtp.style(theme='onedork')
plt.scatter(xhr,yhr,marker = '^')
plt.scatter(xhit, yhit, alpha=.25, marker = 'o')
plt.xlabel("Exit Velocity")
plt.ylabel('Launch Angle')
plt.show()
```

# Model

```python
X = ab_df[['launch_speed','launch_angle']]
y = ab_df['event_binary'].values.reshape(-1,1)

print(X.shape, y.shape)
```

```python
from  sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

# Fit

```python
classifier.fit(X_train, y_train)
```

```python
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

# Predict

```python
predictions = classifier.predict(X_test)
print(f"First 10 predictions: {predictions[:10]}")
print(f"First 10 Actual labels: {y_test[:10].tolist()}")
```
